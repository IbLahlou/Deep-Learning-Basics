## Activation Function

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
x = np.linspace(-5, 5, 100)

plt.style.use('seaborn-whitegrid')
# Set Matplotlib defaults
plt.rc('figure', autolayout=True)
plt.rc('axes', labelweight='bold', labelsize='large',
       titleweight='bold', titlesize=18, titlepad=10)
```

#### 1. Identity Function:

An identity function, also called an identity relation, identity map or identity transformation, is a function that always returns the value that was used as its argument, unchanged. also called persistant model 

**Formula:**

$$id(x) = x $$

#### 2. Binary Step Function:

The Heaviside step function, or the unit step function, usually denoted by H or Œ∏ (but sometimes u, 1 or ùüô), is a step function named after Oliver Heaviside, the value of which is zero for negative arguments and one for positive arguments.

**Formula:**

$$H(x)= 1_{\mathbb{R}_{+}}(x)=\begin{cases} 1 \text{, si  } x \geq 0 \\ 0 \text{, si  } x < 0 \end{cases}$$


```python
def H(x):
    return np.where(x >= 0, 1, 0)
```

```python
plt.plot(x, H(x))
plt.xlabel(r"$x$")
plt.ylabel(r"$H(x)$")
plt.title("Step Function")
plt.text(0.5, 0.8, r"$H(x)= 1_{\mathbb{R}_{+}}(x)$", fontsize=12)
plt.show()
```

#### 3. Sigmoid Function:
The sigmoid function is commonly used in binary classification tasks. It maps the input to a value between 0 and 1, representing the probability of the input belonging to the positive class.

**Formula:**

$$\text{sigmoid}(x) = \frac{1}{{1 + e^{-x}}}$$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
plt.plot(x, sigmoid(x))
plt.title("Sigmoid")
plt.text(0.5, 0.5, r"$\sigma(x) = \frac{1}{1 + e^{-x}}$", fontsize=12)
plt.grid(True)
```

**4. Hyperbolic Tangent (Tanh) Function:**
The hyperbolic tangent function is similar to the sigmoid function but maps the input to a value between -1 and 1. It is often used in classification and regression tasks.

**Formula :**

$$\tanh(x) = \frac{{e^x - e^{-x}}}{{e^x + e^{-x}}}$$

```python
def tanh(x):
    return np.tanh(x)
```

```python
plt.plot(x, tanh(x))
plt.title("Tanh")
plt.text(0.5, 0.5, r"$\tanh(x)$", fontsize=12)
plt.grid(True)
```

**5.Rectified Linear Unit (ReLU) Function:**
ReLU is a popular activation function that maps all negative inputs to zero and keeps positive inputs unchanged. It helps introduce non-linearity to the model and is widely used in deep neural networks.

**Formula :**
$$\text{relu}(x) = \max(0, x)$$

```python
def relu(x):
    return np.maximum(0, x)
```

```python
plt.plot(x, relu(x))
plt.title("ReLU")
plt.text(0.5, 0.5, r"ReLU(x) =$\max(0, x)$", fontsize=12)
plt.grid(True)
```

**6.Leaky ReLU Function:**
Leaky ReLU is a modified version of ReLU that allows small negative values instead of setting them to zero. It helps mitigate the "dying ReLU" problem by addressing dead neurons.

**Formula :**

$$\text{leaky_relu}(x, \alpha) = \begin{cases} \alpha x, & \text{if } x < 0 \\ x, & \text{otherwise} \end{cases}$$

- $\alpha$ represents the negative slope for negative inputs

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
```

```python
plt.plot(x, leaky_relu(x))
plt.title("Leaky ReLU")
plt.text(0.5, 0.5, r"$\alpha*x \ si \ x<0$ ", fontsize=12)
plt.text(0.9, 0.2, r"$\ x \ sinon$ ", fontsize=12)
plt.grid(True)
```

**7. Randomized Leaky ReLU Function:**
Randomized Leaky ReLU is a variation of the Leaky ReLU activation function that introduces randomness to the negative slope. In Randomized Leaky ReLU, the negative part of the input is multiplied by a random value drawn from a uniform distribution instead of using a fixed slope. This introduces additional diversity and flexibility to the activation function, allowing it to adapt better to different types of data.

**Formula:**

$$\text{rand_leaky_relu}(x, \alpha) = \begin{cases} \alpha x, & \text{if } x < 0 \\
x + \epsilon \cdot \text{random_value}, & \text{otherwise}
\end{cases}$$

- $\alpha$ represents the negative slope for negative inputs
- $\epsilon$ is a scaling factor

```python
def randomized_leaky_relu(x, alpha, epsilon):
    return np.where(x >= 0, x, x + epsilon * np.random.uniform(low=0, high=1, size=x.shape))

# Set alpha and epsilon parameters
alpha = 0.1
epsilon = 1

# Generate random values for Randomized Leaky ReLU
np.random.seed(42)
randomized_values = randomized_leaky_relu(x, alpha, epsilon)
```

```python
# Plot Randomized Leaky ReLU
plt.figure(figsize=(6, 4))
plt.plot(x, randomized_values)
plt.title("Randomized Leaky ReLU")
plt.xlabel("x")
plt.ylabel("Randomized Leaky ReLU(x)")
plt.grid(True)
```

**8. Softplus Function:**
The softplus function is a smooth approximation of the ReLU function. It maps the input to a range of positive values and is often used as an activation function for the output layer in regression tasks.

**Formula :**

$$\text{softplus}(x) = \ln(1 + e^x)$$

```python
def softplus(x):
    return np.log(1 + np.exp(x))
```

```python
plt.plot(x, softplus(x))
plt.title("Softplus")
plt.text(0.5, 0.5, r"$\ln(1 + e^x)$", fontsize=12)
plt.grid(True)
```

**9. Swish Function:**
The Swish activation function applies a sigmoid-like transformation to the input, which introduces a non-linearity. It has shown promising performance in some deep learning architectures.

**Formula :**
$$\text{swish}(x) = x \cdot \text{sigmoid}(x)$$

```python
def swish(x):
    return x * sigmoid(x)
```

```python
plt.plot(x, swish(x))
plt.title("Swish")
plt.text(0.5, 0.5, r"$x \cdot \sigma(x)$", fontsize=12)
plt.grid(True)
```

**10. Gaussian Error Linear Unit (GELU) Function:**
GELU is a smooth approximation of the rectifier function. It is used to introduce non-linearity and has been found to perform well in transformer models.

**Formula :**

$$\text{gelu}(x) = 0.5x\left(1 + \tanh\left(\sqrt{\frac{2}{\pi}}\left(x + 0.044715x^3\right)\right)\right)$$

```python
def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3))))
```

```python
plt.plot(x, gelu(x))
plt.title("GELU")
plt.text(0.5, 0.5, r"$\frac{1}{2}x\left(1 + \tanh\left(\sqrt{\frac{2}{\pi}}(x + 0.044715x^3)\right)\right)$", fontsize=12)
plt.grid(True)
```

**11. Exponential Linear Unit (ELU) Function:**

ELU is similar to ReLU but smoothly handles negative inputs by using an exponential function. It helps alleviate the vanishing gradient problem and has negative values for mean activations.

**Formula :**
$$\text{elu}(x, \alpha) = \begin{cases}
x, & \text{if } x \geq 0 \\
\alpha(e^x - 1), & \text{otherwise}
\end{cases}
$$

$\alpha$ is a hyperparameter that controls the value of the negative region slope. 

```python
def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))
```

```python
plt.plot(x, elu(x))
plt.title("ELU")
plt.xlabel(r"$x$")
plt.text(0.7, 0.5, r"$\ \ x \ si \geq 0$", fontsize=12)
plt.text(0.9, 0.2, r"$\alpha(\exp{x} -1)\ sinon$ ", fontsize=12)
plt.ylabel(r"$y$")
plt.grid(True)
```

**12. Softmax Function:**

Softmax is a popular activation function used in multi-class classification tasks. It converts a vector of real numbers into a probability distribution over multiple classes. Softmax ensures that the sum of the probabilities for all classes is equal to 1.

**Formula:**

$$\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{N}e^{x_j}}, \text{ for } i = 1, 2, \ldots, N$$

$x_i$ represents the $i$-th element of the input vector $x$, and $N$ is the number of classes.

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()
```

```python
plt.plot(x, softmax(x))
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.title("Softmax Function")
plt.text(0.01, 0.01, r"$\frac{e^{x}}{\sum_{j=1}^{N}e^{x_j}}$", fontsize=12)
plt.grid(True)
plt.show()
```
